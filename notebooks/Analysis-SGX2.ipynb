```python
import json
import csv
import time
import pandas as pd
from datetime import datetime

import matplotlib.pyplot as plt
import mplfinance as mpf
```

Load file


```python
symbol = 'C09'
TIME, VOLUME, OPEN, HIGH, LOW, CLOSE, ADJ_CLOSE = 0, 1, 2, 3, 4, 5, 6

flatten_data = []
with open(f'../data-dump/{symbol}.SI-max-1mo.json', 'r', encoding='UTF8', newline='') as in_file:
    json_data = json.load(in_file)
chart_json = json_data['chart']
result = chart_json['result'][0]
meta = chart_json['result'][0]['meta']
timestamp = chart_json['result'][0]['timestamp']
indicators = chart_json['result'][0]['indicators']

ohlcv = indicators['quote'][0]
qopen = indicators['quote'][0]['open']
qclose = indicators['quote'][0]['close']
qhigh = indicators['quote'][0]['high']
qlow = indicators['quote'][0]['low']
qvolume = indicators['quote'][0]['volume']
qadjclose = indicators['adjclose'][0]['adjclose']

d_list = []
for i in range(len(timestamp)):
    # print(timestamp[i], volume[i], open[i], close[i], high[i], low[i], adjclose[i])
    d = (
        datetime.fromtimestamp(timestamp[i]), 
        int(qvolume[i]), 
        float(qopen[i]), 
        float(qhigh[i]), 
        float(qlow[i]), 
        float(qclose[i]))
    d_list.append(d)

    # csv_reader = csv.reader(in_file)
    # for row in csv_reader:
    #     flatten_data.append([
    #         datetime.strptime(row[TIME], "%Y-%m-%d %H:%M:%S"), 
    #         int(row[VOLUME]), 
    #         float(row[OPEN]), 
    #         float(row[HIGH]), 
    #         float(row[LOW]), 
    #         float(row[CLOSE])
    #     ])
```

```python
# instrument = json_data['instrument']
# granularity = json_data['granularity']
# candles = json_data['candles']

# print(f"Instrument: [ {instrument} ], Granularity: [ {granularity} ]")
```

```python
# flatten_data = [
#     [ 
#         datetime.strptime(x['time'], "%Y-%m-%dT%H:%M:%S.%f000Z"), 
#         bool(x['complete']), 
#         int(x['volume']), 
#         float(x['mid']['o']), 
#         float(x['mid']['h']), 
#         float(x['mid']['l']), 
#         float(x['mid']['c'])
#     ] for x in candles]
df = pd.DataFrame(
    d_list, 
    columns=['time', 'Volume', 'Open', 'High', 'Low', 'Close'])

# df.index = pd.DatetimeIndex(df['time'])

print(f"df shape: {df.shape}")

# print(df.info())
```

```python
df['ma'] = df['Close'].rolling(20).mean()
df['ewm'] = df['Close'].ewm(span=20, adjust=False).mean()
# df.head(30)
index_df = df.set_index('time')
index_df.tail(28)
```

```python
#min_close_low = round(df.tail(28)[['Open', 'High', 'Close', 'Low']].min(axis=1).min(), 4)
#max_open_high = round(df.tail(28)[['Open', 'High', 'Close', 'Low']].max(axis=1).max(), 4)

support_line = round(df.tail(28)['Close'].min(), 4)
resistant_line = round(df.tail(28)['Close'].max(), 4)
median_line = round(df.tail(28)['Close'].median(), 4)
mean_line = round(df.tail(28)['Close'].mean(), 4)
mode_line = round(df.tail(28)['Close'].mode(), 4)

print(support_line)
print(resistant_line)

# This plots close
mpf.plot(
    index_df.tail(28),
    type='line', 
    volume=True, 
    title=f'{symbol} close plot',
    hlines=dict(hlines=[support_line, resistant_line, median_line, mean_line], colors=['g','r', 'c'], linestyle='-.', linewidths=(1, 1))
    )

# min_close_low = round(df[['Close', 'Low']].min(axis=1).min(), 4)
# max_open_high = round(df[['Open', 'High']].max(axis=1).min(), 4)

# Need someway to auto figure out hlines
mpf.plot(index_df, type='line', 
    mav=(20, 30),
    # hlines=dict(hlines=[min_close_low, max_open_high], colors=['g','r'], linestyle='-.', linewidths=(1, 1)),
    volume=True,
    title=f'{symbol} close plot')
    
```
